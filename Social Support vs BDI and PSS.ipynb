# Does social support help in healing?

This is something that I have been curious about since I started listening to all patient follow up sessions. I saw a lot of cases wherein the patient had a big support group but still had a difficult time coping up with depressive/post-traumatic symptoms. This made me wonder if the nature of the incident was more significant than social support? This is something that I try to explore in this notebook.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
clinical_info_file_path = "clinical_info.xlsx"
df = pd.read_excel(clinical_info_file_path, engine="openpyxl")
```

```python
def visualize_patient_BDI_PSS(patient_id):

    df_patient = df.loc[df['SID'] == patient_id]
    df_BDI = df_patient[['BDItotalscore_ED', 'BDItotalscore_1mo', 'BDItotalscore_3mo', 'BDItotalscore_6mo', 'BDItotalscore_12mo']]
    df_PSS = df_patient[['PSStotal_1month', 'PSStotal_3month', 'PSStotal_6month', 'PSStotal_12month']]
    x_axis = ["ED", "FU-1", "FU-3", "FU-6", "FU-12"]
    df_BDI = df_BDI.values.tolist()[0]
    df_PSS = df_PSS.values.tolist()[0]
    df_PSS.insert(0, None)
    plt.plot(x_axis, df_BDI, label="BDI Scores for Patient "+str(patient_id))
    plt.plot(x_axis, df_PSS, label="PSS Scores for Patient "+str(patient_id))
    plt.legend()
    plt.xlabel('Time')
    plt.ylabel('BDI & PSS Scores')
    plt.title("BDI & PSS Scores for Patient "+str(patient_id))
    plt.show()
```

```python
df_patient_with_zero_social_support = df.loc[(df['ed_social_support_sti14'] == 0)]
patient_list = df_patient_with_zero_social_support['SID'].values.tolist()

for patient in patient_list:
    visualize_patient_BDI_PSS(patient)
```

```python
df_patient_with_one_social_support = df.loc[(df['ed_social_support_sti14'] == 1)]
patient_list = df_patient_with_one_social_support['SID'].values.tolist()

for patient in patient_list:
    visualize_patient_BDI_PSS(patient)
```

```python
df_patient_with_two_social_support = df.loc[(df['ed_social_support_sti14'] == 2)]
patient_list = df_patient_with_two_social_support['SID'].values.tolist()

for patient in patient_list:
    visualize_patient_BDI_PSS(patient)
```

```python
df_patient_with_three_social_support = df.loc[(df['ed_social_support_sti14'] == 3)]
patient_list = df_patient_with_three_social_support['SID'].values.tolist()

for patient in patient_list:
    visualize_patient_BDI_PSS(patient)
```

## Observations, thoughts, and questions

1. Nothing conclusive. Higher social support doesn't show a direct trend with healing.
2. A next step could be to segregate based on type of incidents. This is because it is totally possible that a person who was in a road accident feels better with a support group of 3 people, whereas a sexual assault victim has a harder time in spite of having 3 supportive people around?
3. In cases like these where we see no correlation ourselves, should we still keep pursuing it using some intelligent model (like ML), or can we rule out the possibility of a trend right away?
